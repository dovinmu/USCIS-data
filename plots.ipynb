```python
import pandas
import matplotlib as plt
plt.rc('figure', max_open_warning = 0)
```

```python
df = pandas.read_csv('data/sum_by_year.csv')
```

```python
for state in df.state.unique():
    plot = df[df.state==state].groupby('fy').sum().plot(title=state)
    fig = plot.get_figure()
    fig.savefig('plots/' + str(state) + '.png')
```

```python
plot = df.groupby('fy').sum().plot(title='US nationalization')
fig = plot.get_figure()
fig.savefig('plots/all_US.png')
```
